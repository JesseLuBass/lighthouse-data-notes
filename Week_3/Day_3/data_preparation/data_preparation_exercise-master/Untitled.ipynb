```python
    from sklearn import datasets
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import DBSCAN
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('testdata.csv', index_col = 0)
```

```python
data.drop('Unnamed: 0.1',axis = 1 )
```

```python
iris = datasets.load_iris()
X = iris.data
data2 = pd.DataFrame(X)
```

```python
std_slc = StandardScaler()
X_std = std_slc.fit_transform(X)
```

```python
clt = DBSCAN()

model = clt.fit(X_std)

clusters = pd.DataFrame(model.fit_predict(X_std))
data2["Cluster"] = clusters
```

```python
fig = plt.figure(figsize=(10,10)); ax = fig.add_subplot(111)
scatter = ax.scatter(data2[0],data2[1], c=data2["Cluster"],s=50)
ax.set_title("DBSCAN Clustering")
ax.set_xlabel("X0")
ax.set_ylabel("X1")
plt.colorbar(scatter)
plt.show()
```
