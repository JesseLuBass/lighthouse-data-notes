```python
import pandas as pd
fac = pd.read_csv('Facilities_Database.csv')
```

```python
fac['facdomain'] = fac['facdomain'].astype('category')
```

```python
fac['facdomain_cat']=fac['facdomain'].cat.codes
```

```python
fac.facdomain_cat.dtypes
```

```python
import numpy as np
```

```python
pd.pivot_table(fac, index=['nta'], columns=['facdomain_cat'],values= 'count')
```

```python
fac[['nta','facdomain_cat']].groupby('nta').agg(['count'])
```

```python
onehot = fac[['nta','facdomain']]
```

```python
one_hot = pd.get_dummies(onehot['facdomain'])
```

```python
onehot = onehot.join(one_hot)
```
